```python
# dataset_en = dataset["train"].filter(lambda example: example["language"] == "en")
dataset_en = load_from_disk("data/roots_1e-1_en")
```

```python
len(dataset_en)
```

```python
dataset_en.shuffle()
```

```python
en_slice = pd.DataFrame.from_dict(dataset_en[:10000])
en_slice
```

```python
pd.set_option("max_colwidth", 400)
```

```python
en_slice.groupby("dataset_id")[["doc_id"]].count().reset_index().sort_values("doc_id", ascending=False)
```

```python
en_slice["doc_len"] = en_slice.apply(lambda x: len(x["text"].split()), axis=1)
```

```python
en_slice
```

```python
en_slice["doc_len"].mean()
```

```python
en_slice["doc_len"].max()
```

```python
en_slice_lengths = (
    en_slice.groupby("doc_len")["dataset_id"]
    .count()
    .reset_index(name="count")
    .sort_values("doc_len", ascending=True)
)
```

```python
en_slice_lengths
```

```python
xticklabels = [
    str(en_slice_lengths.iloc[i]["doc_len"]) if i % 100 == 0 else " "
    for i in range(len(en_slice_lengths["doc_len"]))
]
```

```python
ax = en_slice_lengths.plot(
    x="doc_len",
    y="count",
    figsize=(20, 10),
    kind="bar",
)
ax.set_xticklabels(xticklabels)

plt.savefig("myfile.pdf")
```

```python
en_slice_avg = (
    en_slice.groupby("dataset_id")[["doc_len", "text"]]
    .agg({"text": "count", "doc_len": "sum"})
    .reset_index()
    .sort_values("doc_len", ascending=False)
)
```

```python
en_slice_avg["avg_doc_len"] = en_slice_avg.apply(lambda x: x["doc_len"] // x["text"], axis=1)
```

```python
en_slice_avg.sort_values("avg_doc_len", ascending=False).style.set_properties(**{"text-align": "left"})
```
